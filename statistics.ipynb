```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("infoperstudents.csv")
students = data['name'].unique()
profession = data['profession'].unique()
```

```python
print(data.loc[(data["profession"] == "לשון") & (data["name"] == "שקד בן חמו")])
myData = data.loc[(data["profession"] == "לשון") & (data["name"] == "שקד בן חמו")]
print(myData['status'].hist(alpha=.5))
print(myData['status'].describe())
```

```python
import numpy as np
import matplotlib.pyplot as plt
students = data["name"].unique()
students = np.sort(students)
x = ['evasion','present','home']
for student in students: 
    print("student name: ",student)
    myData = data.loc[data["name"] == student]
    evasion = myData.loc[myData['status'] == 'evasion']
    home = myData.loc[myData['status'] == 'home']
    present = myData.loc[myData['status'] == 'present']
    print('amount of evasion:',evasion['status'].count())
    print('amount of home:',home['status'].count())
    print('amount of present:',present['status'].count(), '\n')
    #myData['status'].hist(alpha=.5)
    labels = np.arange(0,56)
    y = [evasion['status'].count(), present['status'].count(), home['status'].count()]
    plt.bar(x, y , width=0.4)
    plt.title(student[::-1])
    plt.show()
    print(myData['status'].describe())
    print('\n\n')

```

```python
import numpy as np
import matplotlib.pyplot as plt
students = data["name"].unique()
professions = data["profession"].unique()
professionslist = []
evasionlist = []
students = np.sort(students)
professions = np.sort(professions)

for student in students: 
    print("student name: ",student )
    myData = data.loc[data["name"] == student]
    evasion1 = myData.loc[myData['status'] == 'evasion']
    evasionStudent = evasion1['status'].count()
    for profession in professions:    
        myDataProfession = myData.loc[(data["profession"] == profession)]      
        evasion = myDataProfession.loc[myData['status'] == 'evasion']
        evasionlist.append(evasion['status'].count())
    if(evasionStudent != 0):      
        for i in professions:
            professionslist.append(i[::-1])  
        plt.bar(professionslist, evasionlist , width=0.4)
        plt.title(student[::-1])
        plt.show()
    professionslist = []
    evasionlist = []
    print(myData['status'].describe())
    print('\n\n')
```
